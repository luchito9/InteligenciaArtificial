al igual que la regresión polinomica es utilizada para predecir un valor continuo, es decir, una variable dependiente que puede tomar cualquier valor dentro de un rango específico.

En la regresión múltiple, el objetivo es modelar la relación entre múltiples variables independientes y una variable dependiente continua.

La regresión múltiple se utiliza para predecir valores continuos y es una herramienta útil en una variedad de campos donde se desea comprender y modelar la relación entre múltiples variables.
